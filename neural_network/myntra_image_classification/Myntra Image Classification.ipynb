```python
import numpy as np
import pandas as pd
import os
import csv
import cv2
import random,string
from PIL import Image
from io import BytesIO
import shutil
from tqdm import tqdm
from sklearn.model_selection import train_test_split 
```

```python
import requests
r = requests.get("https://raw.githubusercontent.com/harishbisht/machine-learning/master/input.csv")

with open(r'input_data.csv', 'wb') as f:
    f.write(r.content)

#shutil.rmtree("raw")
if not os.path.isdir("raw"):
    os.mkdir("raw")
```

```python
def random_word(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))
```

```python
csv_rows = []
with open("input_data.csv","r") as csv_file:
    reader = csv.reader(csv_file)
    next(reader, None)
    csv_rows = filter(None, [row for row in reader if row[4]])

csv_rows = [i for i in csv_rows]

#filter_data
csv_rows = csv_rows[:3000]
```

```python
def random_word(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))


def download_image_and_resize(csv_row,size=(64,64)):
    response = requests.get(csv_row[4])
    img = Image.open(BytesIO(response.content)).convert('L')
    img.thumbnail(size, Image.ANTIALIAS)
    img.save(os.path.join("raw",random_word(10)+ "__" + csv_row[5].replace(" ","_")+".jpeg"))
```

```python
#downloading images through threading

import queue
import threading

q = queue.Queue()

def worker(row_queue):
    queue_full = True
    while queue_full:
        try:
            #get your data off the queue, and do some work
            row = row_queue.get(False)
            download_image_and_resize(row)
            print (str(row_queue.qsize()) + " url left") 
        except queue.Empty:
            queue_full = False

for csv_row in csv_rows:
    q.put(csv_row)

thread_count = 30
for i in range(thread_count):
    t = threading.Thread(target=worker, args = (q,))
    t.start()


# os.listdir("raw")
```

```python
IMG_SIZE = 64

def label_array_generator(sub_category_name):
    sub_category = ['Abstract', 'Biker', 'Camouflage', 'Checked', 'Colourblocked', 'Conversational', 'Floral', 'Geometric', 'Graphic', 'Humour_and_Comic', 'Music', 'People_and_Places', 'Polka_Dots', 'Self_Design', 'Solid', 'Sports', 'Sports_and_Team_Jersey', 'Striped', 'Superhero', 'Tie_and_Dye', 'Tribal', 'Typography', 'Varsity']
    array = [0]*len(sub_category)
    array[sub_category.index(sub_category_name)] = 1
    return array


def fetch_input_label_from_imagename(image_name):
    return label_array_generator(image_name.split("__")[1].split(".")[0])
```

```python
all_images = []

for i in os.listdir("raw"):
    all_images.append(i)
train_list, val_list = train_test_split(all_images, test_size=0.3)
```

```python
def create_model_data(input_list,location="raw"):
    input_data = []
    for img in tqdm(input_list):
        label = fetch_input_label_from_imagename(img)
        path = img
        path = os.path.join(location,img)
        img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMG_SIZE,IMG_SIZE))
        input_data.append([np.array(img),label])
    return input_data


```

```python
train = create_model_data(train_list)
val = create_model_data(val_list)
```

```python
print(len(train))
print(len(val))
```

```python
#Creating training ad Validation arrays
X = np.array([i[0] for i in train]).reshape(-1,IMG_SIZE,IMG_SIZE)
Y = [i[1] for i in train]

val_X = np.array([i[0] for i in val]).reshape(-1,IMG_SIZE,IMG_SIZE)
val_Y = [i[1] for i in val]
```

```python
#Scaling data for neural network
X = X/float(255)
val_X = val_X/float(255)
```

```python
Y = np.asarray(Y)
val_Y = np.asarray(val_Y)
```

```python
X.shape, val_X.shape
```

```python
#Importing keras libraries
import keras
from keras.layers import Input, Dense, Flatten, Dropout
from keras.optimizers import SGD, Adam
from keras.models import Model
from keras.preprocessing import image
from PIL import Image
```

```python
#Network Architecture
x_input = Input(shape=(IMG_SIZE, IMG_SIZE))
x = Flatten()(x_input)
# x = Dense(4096, activation="tanh")(x)
# x = Dropout(0.2)(x)
x = Dense(8096, activation="tanh")(x)
x = Dense(1024, activation="tanh")(x)
x = Dense(256, activation="tanh")(x)
# x = Dense(64, activation="tanh")(x)
x_out = Dense(23, activation="softmax")(x)

#Specifying input and output
model = Model(inputs=x_input, outputs=x_out)
#sgd = SGD(lr=0.02, momentum=0.0, decay=0.0, nesterov=False)
adam = Adam(lr=0.00001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(optimizer=adam, loss = "binary_crossentropy", metrics=["accuracy"])
model.summary()

history = model.fit(x=X, 
                    y=Y, 
                    validation_data=(val_X, val_Y), 
                    epochs = 40, 
                    batch_size=128, 
                    verbose=1)

import matplotlib.pyplot as plt
prediction = model.predict(val_X)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

score = model.evaluate(val_X, val_Y, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
